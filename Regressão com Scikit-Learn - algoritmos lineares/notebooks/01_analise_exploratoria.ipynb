```python
from sklearn.datasets import load_diabetes
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
dados = load_diabetes(as_frame=True, scaled=False)

dados
```

Para pegar os dados use `dados["frame"]`

```python
df = dados['frame']

df.head()
```

```python
novos_nomes = {"age": "idade", "sex": "sexo", "bmi": "imc", "bp": "pressao_media", "s1": "colesterol_total", "s2": "ldl", "s3": "hdl", "s4": "colesterol_hdl", "s5": "triglicerides", "s6": "glicose", "target": "target"}
df = df.rename(columns=novos_nomes)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
corr = df.corr(numeric_only=True)

fig = px.imshow(corr, text_auto=True, color_continuous_scale="RdBu_r", zmin=-1, zmax=1, template="plotly_dark")
fig.update_layout(width=1000, height=800, margin=dict(l=80, r=80, t=80, b=80))
fig.show()

```

A coluna `colesterol_hdl` parece que está arrendondada e isso pode trazer maleficios ao modelo.

```python
#vamos ver quantos numeros são inteiros

df['colesterol_hdl'].apply(lambda x: x.is_integer()).value_counts()

#ou seja 380 são inteiros que representa 86%
```

```python
serie = df["colesterol_hdl"].apply(lambda x: x.is_integer()).value_counts()
percentual_inteiros = (serie.get(True, 0) / serie.sum()) * 100

fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=percentual_inteiros,
    title={"text": "Colesterol/HDL com valor inteiro (%)"},
    gauge={
        "axis": {"range": [0, 100]},
        "bar": {"thickness": 0.3},
        "steps": [
            {"range": [0, 50]},
            {"range": [50, 80]},
            {"range": [80, 100]}
        ]
    }
))

fig.update_layout(template="plotly_dark")
fig.show()
```

Talvez seja necessário fazer um tratamento nessa coluna. Pq como ela tem alguns dados categoricos e mistura dados não categoricos

### Análises do boxplots

```python
cols = df.select_dtypes(include="number").columns

fig = px.box(df[cols])
fig.update_layout(template="plotly_dark")
fig.show()
```

Otimizando tipos

```python
#trocando de floats para inteiros

colunas_valores_inteiro = []

for coluna in df.columns:
    if df[coluna].apply(lambda valor: valor.is_integer()).all():
        colunas_valores_inteiro.append(coluna)


colunas_valores_inteiro
```

```python
df[colunas_valores_inteiro] = df[colunas_valores_inteiro].astype(int)

df.info()
```
